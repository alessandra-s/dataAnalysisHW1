#### 1. Write a function that checks if a string is a palindrome or not. This function should output a Boolean. (5 pts)

For this first iteration do it with a loop

```python
# For Loop Implementation 
#isPalindrome takes in string input s

def isPalindrome (x):
   #if x has spaces remove them
    x = x.replace(" ", "")
    #make all letters lowercase
    x=x.lower()
    
    # store last entry in x as a variable
    lengthX= len(x)
    

    #for loop iterates through half the length of the word (floored)
    # // = floor divison
    for i in range(lengthX//2):

        #compare 1st half of word to latter half of word
        if x[i] == x[-1-i]:
            isPalindrome = True 
        else:
            isPalindrome = False
    
    return isPalindrome

x= 'plo'
p= "jokfdskfdfjkl"
owl ='Mr Owl ate my metal worm'

print(isPalindrome(x))
print(isPalindrome(p))
print(isPalindrome(owl))
```

```python
## Implement 
#isPalindrome takes in string input of x

def palindromeTest (x):
#if x has spaces remove them
    x = x.replace(" ", "")
#make all letters lowercase
    x=x.lower()

#take input word and store it backwards using slicing
    lengthX= len(x)
    xBackwards= x[lengthX::-1]
#compare x & xBackwards to see if the same, return true if yes
    if x == xBackwards:
        palindromeTest = True
    else:
        palindromeTest = False
    return palindromeTest


test1= 'A nut For a jar of tuNa'
test2= 'noon'
    
print(palindromeTest(test1))
print(palindromeTest (test2))
```

You're on the medal handing out committee for field day, and you are in charge of giving out the silver medal. The only problem is, there are supposed to be too many runners to count. Write a function that takes in a list of scores and outputs the runner up score. 

```python
def findRunnerUp (x):
    #find the 2nd place winner in a score list
    #1st place= highest score
    
    #sort list descending order
    x.sort(reverse=True)
    #store 2nd entry as silver medal winner
    silverMedal = x[1]
    
    return silverMedal
```
